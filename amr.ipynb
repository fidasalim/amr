```python
#
import numpy as np
import matplotlib.pyplot as plt
import math as m

def create_matrix(N,a,b):
  A = np.zeros((N,N))
  for i in range(0,N-1):
    A[i,i+1] = b
    A[i,i] = a
    if (i>0):
      A[i,i-1] = b
  A[N-1,N-2] = b
  A[N-1,N-1] = a
  return(A)

def y_matrix(U):
    import math as m
    h = 1.0546e-34
    mass = 9.11e-31
    f1 =  lambda x: ((2*mass/h**2)**(3/2))*m.sqrt(abs(x-10))/(2*(m.pi**2))
    X = U
    Y = [f1(x) for x in X]
    return(Y)

def graph_plot(X,Y,N):
    fig, ax = plt.subplots()
    ax.stem(X, Y, markerfmt=' ')
    plt.title('N ='+str(N))
    plt.savefig('fig'+str(N)+'.png')
    plt.show()

for N in range(1,11):
  A = create_matrix(pow(2,N),10,-1)
  results = np.linalg.eig(A)
  Y = y_matrix(results[0])
  graph_plot(results[0],Y,pow(2,N))
```

```python
#
import numpy as np
import matplotlib.pyplot as plt
import math as m

def create_matrix(N,a,b,c):
  A = np.zeros((N,N))
  for i in range(0,N-1):
    if i%2 == 0:
      A[i,i+1] = b
    else:
      A[i,i+1] = c
    A[i,i] = a
    if (i>0):
      if i%2 != 0:
        A[i,i-1] = b
      else:
        A[i,i-1] = c
  if N %2 == 0:
    A[N-1,N-2] = b
  else:
    A[N-1,N-2] = c
  A[N-1,N-1] = a
  return(A)

def y_matrix(U):
    import math as m
    h = 1.0546e-34
    mass = 9.11e-31
    f1 =  lambda x: ((2*mass/h**2)**(3/2))*m.sqrt(abs(x-10))/(2*(m.pi**2))
    X = U
    Y = [f1(x) for x in X]
    return(Y)

def graph_plot(X,Y,N):
    fig, ax = plt.subplots()
    ax.stem(X, Y, markerfmt=' ')
    plt.title('N ='+str(N))
    plt.savefig('fig'+str(N)+'.png')
    plt.show()

U = [2,9,65,146,347,568,1023,2000]
for N in U:
  A = create_matrix(N,10,-1,-2)
  results = np.linalg.eig(A)
  Y = y_matrix(results[0])
  graph_plot(results[0],Y,N)
```
